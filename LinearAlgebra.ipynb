## Linear Algebra
1. Matrix Addition and Scalar Multiplication

```python
%matplotlib inline 
# Importing the library
import numpy as np
import matplotlib.pyplot as plt # For plotting
```

```python
# Definig 2 square matrix 
# Using the numpy's matrix 
A = np.matrix('1 2; 3 4')
B = np.matrix('1 0; 2 5')
```

```python
print A
```

```python
C = A + B
print C
```

**Visualizing the added matrix**

```python
O = [0, 0]
plt.figure(1, figsize=(12, 4))
plt.subplot(131) # numbers are in the order row, column, figure number
plt.quiver(O, O, [1, 3], [2, 4], angles='xy', scale_units='xy', scale=1, color=['r','b'])
plt.grid(True)
plt.xlim(-0.1, 10)
plt.ylim(-0.1, 10)
plt.title('Matrix plane A')

plt.subplot(132) #numbers are in the order row, column, figure number
plt.quiver(O, O, [1, 0], [2, 5], angles='xy', scale_units='xy', scale=1, color=['r','b'])
plt.grid(True)
plt.xlim(-0.1, 10)
plt.ylim(-0.1, 10)
plt.title('Matrix plane B')

plt.subplot(133) #numbers are in the order row, column, figure number
plt.quiver(O, O, [2, 2], [5, 9], angles='xy', scale_units='xy', scale=1, color=['r','b'])
plt.grid(True)
plt.xlim(-0.1, 10)
plt.ylim(-0.1, 10)
plt.title('Matrix C = Matrix plane A + B')
plt.show()
```
