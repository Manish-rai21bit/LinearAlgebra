y2 = a21*x1 + a21*x2 <br>

**_This set of equation may relate an x1 x2-coordinate system to y1 y2-coordinate system in a plane._** <br>

In **Vectorial Form** this equation is written as : <br>
y = [y1 y2]T ; A = [[a11 a12][a21 a22]]; x = [x1 x2]T , Where T is for transpose <br>
y = Ax

```python
# Q. Solving a Markov Process to get the Area in use by various sectors
# Matrix A is a Stochastic matrix (driven by probability) showing the transition form one state to other
# A is called the transition Matrix
A = np.matrix('0.7 0.1 0;\
              0.2 0.9 0.2;\
              0.1 0 0.8')
IA = np.matrix('25 20 55') #initial area/state under Commetrial use, Industrial Use, Residential Use
```

```python
FS1 = np.matmul(A, np.transpose(IA))
FS2 = np.matmul(A, FS1)

print 60*FS1, '\n', 60*FS2
```

```python
# Q2. If the transition Matrix and Initial State are :
A = np.matrix('0.5 0.3; 0.5 0.7')
IS = np.matrix('1; 1')
# Whats the next 3 states?

FS1 = np.matmul(A, IS)
FS2 = np.matmul(A, FS1)
FS3 = np.matmul(A, FS2)

print FS1, '\n',FS2, '\n', FS3
```

### Linear Independence, Rank of a Matrix. Vector Space
After applying the transformation to an identity matrix, the maximum number of dimensions remaining is called the rank of a matrix. In tech term is is the maximum number of row vectors of A which are linearly independent.

* If m-Vectors are linearly combined and the Linear Combination value is 0 only when the scalars are 0, then the vectors are Linear Independent<br>
* Another way to find if a set of vectors is Linearly independent is by having a matrix with the vectors as rows. If the rank of matrix = number of rows then it's linearly independent.
* Rank of a matrix can be (Visualized) or be found using the Row equivalence matrix. <br>
* Rank of a matrix can be (also) found by using the Column equivalence matrix.

```python
np.rank(FS1)
```

```python
def get_linear_independent_func(matrix):
    if np.rank(matrix) == 4:
        C = "Independent"   
    else :
        C = "Dependent"
    return C
```

```python
# Q. Find if the following set of vectors is Linearly independent
a1 = [3, -2, 0, 4]
a2 = [5, 0, 0, 1]
a3 = [-6, 1, 0, 1]
a4 = [2, 0, 0, 3]
A = np.matrix('3, -2, 0, 4; 5, 0, 0, 1; -6, 1, 0, 1; 2, 0, 0, 3')
print np.rank(A), '\n', get_linear_independent_func(A)
```

#### Vector Space:
A **vector space** is a (nonempty) set **V** of vectors with the same number of components such that with ny two vectors **a** and **b** in V all their linear combinations are elements of V<br>

#### Solution of Linear Systems: Existence, Uniqueness
1. Existence - solution exists if rank of _coefficient matrix A_ and the _augmented matrix A'_ has the same rank.
2. Uniqueness - solution is unique if the ranks of the _coefficient matrix A_ and the _augmented matrix A'_ has the same and equal to the n - number of unknowns.

### Inverse of a Matrix. & Gauss-Jordan Elimination
* We consider Square Martix here
* If **A** has an inverse, then **A** is nonsingular matrix. If **A** has no inverse, then **A** is Singular matrix. 
* **A** has an inverse only if it has the maximum possible rank (This can be thought of geometrically)<br>
- if the rank is less than maximum, which means the dimensions collapses and if inverse applied to the collapsed dimensional space.. We won't be able to get to the identity matrix.
