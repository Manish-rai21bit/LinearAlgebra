```python
# Find the components of the vector v with initial point P and terminal point Q.
# Find |v| and sketch it

def get_componet_vector_func(A, B):
    return list(A[i] - B[i] for i in range(len(A)))
```

#### Inner Products (Dot Products) & Vector Products (Cross Products)
- Dot product of 2 vectors is a scalar quantity and is given as:
    - a.b = |a||b|Cos(ø)
    or 
    - if a = [a1, a2, a3] and b = [b1, b2, b3], then <br>
    a.b = a1b1 + a2b2 + a3b3 <br>
    - Dot products is applicable in cases like: <br>
    Work = Force(in the direction of the movement) x Displacement, and similar cases. Here _work_ is scalar.
- Vector Product (also known as Cross product or Outer product) gives a vector governed by the right hand thumb rule. The magnitude is the area under the parallogram covered by the 2 vectors.

```python
# Dot product of 2 vectors is a Scalar
np.dot([1, 2], [3, 2])
```

```python
# Cross product of 2 vectors is a vector in the direction governed by the right hand thumb rule
np.cross([1, 2], [3, 2])
```

**Parametric representation of a curve and surfaces**. <br>
The field of vector calculus concerned with _curves_ and _surfaces_ and their physics and geometry is called **differential geometry**.<br>
The curve C in space looks as a path of a moving body in space and can be written in a **Parametric representation** as below:
\begin{equation}\ r(t) = [x(t), y(t), z(t)] = x(t)i + y(t)j + z(t)k \end{equation}

Tangent to a curve: <br>
The tangent/slope of a curve can be derived using the conventional way of (height/base) and the general purpose formula can be written down as:
\begin{equation}\ r'(t) = lim_{\Delta t -> 0} \frac{1}{\Delta t}[r(t + \Delta t) - r(t)] \end{equation}

And, the unit vector of the tangent can be written as:
\begin{equation}\ u = \frac{r'}{|r'|} \end{equation}


**Mean Value Theorems** with a very simple explanation in this [video](https://www.youtube.com/watch?v=4Q37iOyBq44) <br>
So what I understand from the MVT is that given a curve. Say we travelling from point A to point B, with time on the x-axis and distance in the y-axis, then at some point (c) in the journey we'd have been at the average speed. The formula for the MVT can be mathematically written as:
\begin{equation}\ f'(c) = \frac{f(b) - f(a)}{b - a} \end{equation}
In a Vector Calculus sense, let f(x, y, z) be continuous and have a continuous first partial derivativein the domain D in the xyz-space. <br>
Let P : f(x, y, z) and Q : f(x+h, y+k, z+l) in the domain D such that the straight line segment PQ joining thes points lie entirely in D. Then,
\begin{equation}\ f(x+h, y+k, z+l) - f(x, y, z) = h \frac{\partial f}{\partial x} + k \frac{\partial f}{\partial y}  + l \frac{\partial f}{\partial z} \end{equation}

#### Gradient of a Scalar Field. Directional Derivative
The **Gradient** of a scalar function f(x, y, z) is denoted by grad f or nabla f and is a vector and is defined as:
\begin{equation}\ grad\text{ } f = \nabla f = [\frac{\partial f}{\partial x}, \frac{\partial f}{\partial y}, \frac{\partial f}{\partial z}] = \frac{\partial f}{\partial x} i + \frac{\partial f}{\partial y} j + \frac{\partial f}{\partial z} k\end{equation}
