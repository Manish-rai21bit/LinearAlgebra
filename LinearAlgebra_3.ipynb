**ijk Notation for vectors** <br>

\begin{equation}\text{vector } a = [a_1, a_2, a_3] = a_1i + a_2j + a_3k\text{ ---- forms a real vector space }R^3\end{equation}
The vector space has a dimension 3. Triple of vectors **i, j, k** is called a standard bias of R^3

```python
# Find the components of the vector v with initial point P and terminal point Q.
# Find |v| and sketch it

def get_componet_vector_func(A, B):
    return list(A[i] - B[i] for i in range(len(A)))
```

```python
A = (3, 2, 0)
B = (5, -2, 0)
get_componet_vector_func(A, B) # Component of a vector
```

```python
np.linalg.norm(get_componet_vector_func(A, B)) # norm or |v| or eucledian distance
```

```python
A = (1, 1, 1)
B = (-4, -4, -4)
get_componet_vector_func(A, B), np.linalg.norm(get_componet_vector_func(A, B))
```

#### Inner Products (Dot Products) & Vector Products (Cross Products)
- Dot product of 2 vectors is a scalar quantity and is given as:
    - a.b = |a||b|Cos(ø)
    or 
    - if a = [a1, a2, a3] and b = [b1, b2, b3], then <br>
    a.b = a1b1 + a2b2 + a3b3 <br>
    - Dot products is applicable in cases like: <br>
    Work = Force(in the direction of the movement) x Displacement, and similar cases. Here _work_ is scalar.
- Vector Product (also known as Cross product or Outer product) gives a vector governed by the right hand thumb rule. The magnitude is the area under the parallogram covered by the 2 vectors.

```python
# Dot product of 2 vectors is a Scalar
np.dot([1, 2], [3, 2])
```

```python
# Cross product of 2 vectors is a vector in the direction governed by the right hand thumb rule
np.cross([1, 2], [3, 2])
```

```python
# Scalar fields
# 1. Determine the isotherms of temperature fields for the scalar functions.
# T = xy ; 
T = random.randint(5, 10)
x = sorted(list(random.uniform(0, 5) for i in range(100))) #this is called generator
y = list(T/x[i] for i in range(100))
```

```python
plt.plot(x,y)
plt.xlim(0, 5)
plt.ylim(0, 50)
plt.grid()
plt.show()
```

```python
# 1. Determine the isotherms of temperature fields for the scalar functions.
# T = 4x - 3y ; 
T = random.randint(5, 10)
x = sorted(list(random.uniform(0, 5) for i in range(100))) #this is called generator
y = list((4*x[i] - T)/3 for i in range(100))

plt.plot(x,y)
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.grid()
plt.show()
```

```python
# Scalar field in Space
# F = x^2 + y^2 + 4z^2 = Constant
F = random.randint(5,10)
x = sorted(list(random.uniform(-5, 5) for i in range(10000)))
y = list(random.uniform(-5, 5) for i in range(10000))
z = list((F - x[i]*x[i] - y[i]*y[i])/4 for i in range(10000))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z)

plt.show()
```
